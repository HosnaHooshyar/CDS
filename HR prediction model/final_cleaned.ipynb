Before starting analysis, it could be useful to look at the distribution of the target among the different features. This is also a good way to visualize our dataset for being aware of any unbalancement that may influence our analysis.

```python
def plotPercentageByTarget(data, column, x_label=None, x_ticks=None):
    
    '''
    Take as input the dataframe and one column as string and print the bar plot
    of percentage of that column in the dataframe according to the target.
    In addition it takes the name to give to x axis and the labels for x ticks
    
    '''
    
    #create percentages
    temp = pd.DataFrame(data[[column, 'target']].groupby([column, 'target'])['target'].count())
    temp = temp.rename(columns={'target':'perc'})
    temp['perc'] = (temp['perc']*100)/len(data)
    temp = temp.reset_index().rename(columns={column:'column', 'target':'target'})
    temp = pd.DataFrame(temp)
    temp['target'] = temp['target'].replace([0.0, 1.0], 
                                            ["Do not look for a new job", "Look for a new job" ])
    
    #plot
    fig = plt.figure(figsize=[8,5])
    sns.barplot(x='column', y='perc', hue='target', alpha=1, data=temp)
    
    #labels
    if x_label is None:
        x_label = column.capitalize()
        
    plt.title("Percentage of {} by target".format(x_label.lower()), fontsize=16)
    
    plt.xlabel(x_label, fontsize=14)
    plt.xticks(fontsize=12)
    if x_ticks is not None:
        plt.xticks(ticks=range(0,len(x_ticks),1), labels=x_ticks)
        
    plt.ylabel("%", fontsize=14)
    plt.yticks(fontsize=12)
    
    plt.legend(title="Target", title_fontsize=14, fontsize=12, labelspacing= 0.8, 
               borderpad=0.9, shadow=True, loc='best')
```

## Preprocessing

Data preprocessing is an integral step in machine learning as the quality of data and the useful information that can be derived from it directly affects the ability of our model to learn; therefore, it is extremely important that we preprocess our data before feeding it into our model.

To do so, we chose two distinct ways:

- Information Gain: by using this preprocessing algorithm we choose features that give the biggest number of information. Thus, it is a process that *selects* existing features;
- PCA and FAMD: this algorithm creates different features basing on the existing ones.

This two ways give us the possibility to see the differences between *select* and *create* according to the accuracy of the prediction in classification algorithms.

### How handle categorical values?

Since we found different methods of doing that, each one with its pros and cons, we decided to choose two of them and analyze the differences that this choice may have caused. These two methods are:
* Dummy values
* Label encoding

Although it was said that these two methods will be used, the columns and their values show that it may not be needed for all of them. Some of these features have a numerical nature and their values can be replaced by some modifications. For example, the values of the *company_size* column are ranges of numbers that can be simply replaced by the mean of the range.

```python
def replace_fun(dataFrame, column, current_value, new_value):
    
    '''
    column is the name of the column that we want to replace its values.
    '''
    
    dataFrame[column] = np.where(
        dataFrame[column] == current_value, new_value, dataFrame[column])
```

```python
def convert_to_float(dataFrame, column):
        dataFrame[column] = dataFrame[column].astype(float)
```

### Information Gain

Information gain can be used for feature selection, by evaluating the gain of each variable in the context of the target variable. In this usage, the calculation is referred to as mutual information between the two random variables.

```python
def infoGain (y_class,X_class,i,j):
    
    '''
    y_class: The value that we want to predict
    X_class: Predictors
    i and j: size of plot
    
    '''
    importance = mutual_info_classif(X_class, y_class)
    feature_importance = pd.Series(importance, index=X_class.columns)
    fig,ax = plt.subplots()
    fig.set_size_inches(i,j)
    ax.barh(feature_importance.sort_values().index, feature_importance.sort_values().values)
    plt.xticks(rotation=90)
    plt.title('Information Gain for each feature')
    plt.show()
    index_list = list(feature_importance.sort_values().index)
    value_list = list(feature_importance.sort_values().values)
    data_list = [index_list,value_list]
    dl = pd.DataFrame (data_list)
    dl = dl.T
    dl.columns = ['feature', 'info_gain']
    dl = dl.set_index('feature')
    dl.sort_values('info_gain', axis = 0, ascending = False,inplace = True)
    dl = dl.drop(dl[dl.info_gain < 0.005].index)
    plt.savefig('infogain.png')
    return dl
```

### Feature Scaling

Feature scaling is mapping the feature values of a dataset into the same range. Feature scaling is crucial for machine learning algorithms that consider distances between observations because the distance between two observations differs for non-scaled and scaled cases. There are different methods for feature scaling, including:
* <b>Normalization</b>: maps the values into the $[0, 1]$ interval

$$z = \dfrac{x-\min(x)}{\max(x) - \min(x)} $$

* <b>Standardization</b>: shifts the feature values to have a mean of zero, then maps them into a range such that they have a standard deviation of 1

$$z = \dfrac{x-\mu}{\sigma} $$

Standardization is more flexible to new values that are not yet seen in the dataset. That is why it is preferred in general.
[Link](https://www.baeldung.com/cs/svm-feature-scaling#:~:text=Feature%20scaling%20is%20crucial%20for,non%2Dscaled%20and%20scaled%20cases.&text=Hence%2C%20the%20distance%20between%20data,the%20decision%20boundary%20SVM%20chooses.)

```python
#scaling function

def scaling(data):
    
    X = data.drop(labels = "target", axis = 1)
    y = data["target"]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=109)

    scaler = StandardScaler().fit(X_train)
    X_std = scaler.transform(X)
    
    scaled_data = pd.DataFrame(X_std, columns=X.columns)
    scaled_data['target'] = list(y)
    
    return scaled_data
```

### Neural Network

The main problem of building a neural network is, for sure, deciding how many hidden layers it should contain and, in addition, how many neurons a single layer should have. To solve this issue, we looked at different opinions in internet and we found an interesting hypotesys: **the majority of the problems can be solved witha single hidden layer; it is rare to improve the accuracy by adding additional layers**.

In the code below we tried to empirically prove this statement with ad additional condition: **to avoid overfitting, the number of neurons in hidden layer ($N$) should satisfy**

$$N \leq \dfrac{N_s}{(\alpha*(N_i+N_o))}$$

where

$N_s$ = number of samples in training data set;

$\alpha$ = an arbitrary scaling factor usually 2-10;

$N_i$ = number of input neurons;

$N_o$ = number of output neurons.


```python
# Function to create model, required for KerasClassifier
def create_model(input_dim, neurons=1, activation='sigmoid', dropout_rate='0.0'):
    # create model
    model = keras.Sequential()
    model.add(keras.layers.Dense(neurons, input_dim=input_dim, activation=activation))
    model.add(keras.layers.Dropout(dropout_rate))
    model.add(keras.layers.Dense(1, activation=activation))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
def BuildNN(input_shape, num_layers, num_neurons_per_layer, dropout_rate, activ_fun):
    
    '''
    
    This function builds a neural network. It takes as input:
        - input_shape = shape of data to be processed as tuple (n,1) where 
                        n:number of features
        - num_layers = number of hidden layers for the neural network 
        - num_neurons = a list of ints which specifies how may neurons each layer should have.
                        The lenght of the list should be the number of layers in the NN
        - activ_fun = the function used for running the NN. Possible values are {'sigmoid', 
                        'relu', 'tanh'}
        - loss_fun = the function used for computing the loss. It should be chosen according
                     to the number of neurons in the output
    
    '''
    

    #build the neural network
    model = keras.Sequential([keras.layers.Flatten(input_shape=input_shape)])
    
    #hidden layers
    for layer in range(num_layers):
        model.add(keras.layers.Dense(num_neurons_per_layer[layer], activation=activ_fun))
        model.add(keras.layers.Dropout(dropout_rate))
    
    #output layer    
    model.add(keras.layers.Dense(1, activation=activ_fun))

    #compile model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics =['accuracy'])
    
    return model
```

### Difficulties and Future Works

The main difficulties we found can be grouped into two categories: hardware and software. Unfortunately, our computers were not able to handle some of the algorithms. For this reason, for example, we cannot use SVM for Information Gain dataset or use cross validation with cv more than 3. From a software point of view, we noticed that for every decision we had to make, there were too many options that were almost the same and that made the selection process really challenging. Another main problem was deu to the dataset itself. Indeed, there was no high correlation among the features and the features that were supposed to be related to the target (like training hours) were not so important. Additionally, the umbalancement of the dataset made our predictions biased on the most common target.

One analysis that we wanted to do was looking for gender discrimination into our dataset. However, the percentage of females candidate was really low (6,5%) thus, this dataset was not the best choice for such an analysis. We wanted to do a gender analysis related to a domain which is not domestic violence or fashion but most of the datasets we found care about gender in this context. One more appropriate dataset would have been a study of salary of data scientists where the gender is specified. We were really happy to extend our work in a topic like that.
